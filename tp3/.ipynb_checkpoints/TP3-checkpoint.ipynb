```python
import matplotlib.pyplot as plt
import numpy as np

from Bio import SeqIO
```

```python
def matriz(seq1, seq2):
    m = np.zeros((len(seq1), len(seq2)))
    
    for i in range(len(seq1)):
        for j in range(len(seq2)):
            if seq1[i] == seq2[j]:
                m[i][j] = 1

    return m

matriz("ACCG", "ATCA")
```

```python
def dot_plot_texto(seq1, seq2):
    print("  " + " ".join(seq2))
    for i, fila in enumerate(matriz(seq1, seq2)):
        print(seq1[i], end =" ")
        for j, col in enumerate(fila):
            print("x" if col else "-", end=" ")
        print()
        
dot_plot_texto("ACCG", "ATCA")
```

```python
def dot_plot(seq1, seq2):
    figure = plt.figure(figsize=(15, 15)) 
    
    axes = figure.add_subplot(111)
    caxes = axes.matshow(matriz(seq1, seq2), cmap="Greys") 
    axes.set_yticklabels([""]+list(seq1)) 
    axes.set_xticklabels([""]+list(seq2)) 

    return axes

dot_plot("ACCG", "ATCA")
```

```python
def words(seq, n):
    return [seq[i:i+n] for i in range(0, len(seq), n)]
```

```python
from Bio.SubsMat import MatrixInfo

def dot_plot_scores(seq1, seq2, sub_mat):
    ax = dot_plot(seq1, seq2)
    
    for i in range(len(seq1)):
        for j in range(len(seq1)):
            ax.text(i, j, sub_mat[seq1[i], seq2[j]], ha="center", va="center", color="red")

    return ax

blosum62 = {}
for k, v in MatrixInfo.blosum62.items():
    blosum62[k] = v
    blosum62[(k[1], k[0])] = v

dot_plot_scores("MGSKC", "MWSKC", blosum62)
```

```python
def alinear(seq1, seq2, sub_mat, gap_penalty):
    ini = np.zeros((len(seq1)+1, len(seq2)+1))
    
    for i in range(1, ini.shape[0]):
        ini[i][0] = i*gap_penalty
    
    for j in range(1, ini.shape[1]):
        ini[0][j] = j*gap_penalty
    
    
#     print(ini)
    
    top = np.copy(ini)
    left = np.copy(ini)
    diag = np.copy(ini)
    score = np.copy(ini)
    
    for i in range(1, ini.shape[0]):
        for j in range(1, ini.shape[1]):
#             s = 2 if seq1[i-1] == seq2[j-1] else 0
            s = sub_mat[(seq1[i-1], seq2[j-1])]
            diag[i][j] = diag[i-1][j-1] + s
            top[i][j] = score[i][j-1] + gap_penalty
            left[i][j] = score[i-1][j] + gap_penalty
            score[i][j] = max(diag[i][j], top[i][j], left[i][j])

    print(score)
    print()
    print(diag)
    print()
    print(top)
    print()
    print(left)
    print()
    
    dirs = []
    i, j = ini.shape[0]-1, ini.shape[1]-1
    while i != 0 and j != 0:
        scores = [
            diag[i][j],
            top[i][j],
            left[i][j],
        ]
        
        dir = scores.index(max(scores))
        print(dir)
        
        if dir == 1 or dir == 0:
            j -= 1
        if dir == 2 or dir == 0:
            i -= 1
        
        dirs.append(dir)
    return dirs


def dna_sub_mat(match, mismatch):
    sub_mat = {}
    for b1 in "ATCG":
        for b2 in "ATCG":
            sub_mat[(b1, b2)] = 2 if b1 == b2 else 0
    return sub_mat

seq1 = "TCGAGA"
seq2 = "TAAGAC"

dirs = alinear(seq1, seq2, dna_sub_mat(2, 0), -1)
```

```python
def traza(seq1, seq2, dirs, dot_plot):
    i, j = len(seq1)-1, len(seq2)-1
    
    out_seq1 = ""
    out_seq2 = ""
    for dir in dirs:
        if dir == 0:
            out_seq1 = seq1[i] + out_seq1
            out_seq2 = seq2[j] + out_seq2
            dot_plot.plot([i,i-1],[j,j-1], color="red", linewidth="5")
            i -= 1
            j -= 1
        
        if dir == 1:
            out_seq1 = "-" + out_seq1
            out_seq2 = seq2[j] + out_seq2
            dot_plot.plot([i,i],[j-1,j], color="red", linewidth="5")
            j -= 1
    
        if dir == 2:
            out_seq1 = seq1[i] + out_seq1
            out_seq2 = "-" + out_seq2
            dot_plot.plot([i-1,i],[j,j], color="red", linewidth="5")
            i -= 1
    print(out_seq1)
    print(out_seq2)
    return dot_plot

dp = dot_plot(seq1, seq2)
traza(seq1, seq2, dirs, dp)
```
